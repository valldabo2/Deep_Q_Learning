```python
from DQN import train, DQNAgent
import gym

def phi(state):
    return state.reshape(1, -1)
env_id='LunarLander-v2'
env = gym.make(env_id)
state_shape = env.observation_space.shape
action_space = env.action_space

layers = [256,256]
activation = 'relu'
start_eps = 1
end_eps = 0.01
eps_steps = 50000
optimizer = 'rmsprop'
loss='huber'
lr = 0.0025
mem_capacity = 200000
batch_size = 128
gamma = 0.99

# When using double dqn
update_target_network_freq = 1000
double_dqn=True
dueling=True

#Per
per=True
per_alpha = 0.7
per_is = True
per_beta_start = 0.5
per_beta_steps = 50000


print_ep_freq = 1
train_steps = 500000
train_freq = 10
warmup = 5000

runs = 5
results = []

for i in range(runs):
    print('Run_'+str(i))
    agent = DQNAgent(state_shape, action_space, layers, activation,
                     start_eps, end_eps, eps_steps, optimizer, loss,lr,
                     mem_capacity, batch_size, gamma, double_dqn,
                     update_target_network_freq, per, per_alpha, dueling, warmup,
                    per_is, per_beta_start, per_beta_steps)

    q_values, rewards, steps = train(agent, env, train_steps, phi, train_freq, warmup, print_ep_freq)
    
    results.append((q_values,rewards,steps))
```

```python
%matplotlib inline
import pandas as pd
import numpy as np

results_df = pd.DataFrame([np.nan]*train_steps, columns=['temp'])
for i,res in enumerate(results):
    temp = pd.DataFrame({'Run_'+str(i):res[1]}, index=res[2])
    results_df = results_df.join(temp)
results_df.index.name = 'train_step'
_ = results_df.pop('temp')
```

```python
results_df.interpolate().rolling(10).mean().mean(axis=1).plot();
```

```python
filename = env_id + '_double_' + str(double_dqn) + '_per_' + str(per) + '_duel_' + str(dueling) + '.csv'
results_df.to_csv('rewards_'+filename)
```

```python
Q_df = pd.DataFrame([np.nan]*train_steps, columns=['temp'])
for i,res in enumerate(results):
    temp = pd.DataFrame({'Run_'+str(i):res[0]}, index=np.arange(len(res[0])))
    Q_df = Q_df.join(temp)
Q_df.index.name = 'train_step'
_ = Q_df.pop('temp')
```

```python
Q_df.interpolate().rolling(10).mean().mean(axis=1).plot();
```

```python
Q_df.to_csv('Q_'+filename)
```
